```python
%matplotlib nbagg
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as LA
```

```python
data_samples = np.linspace(-2,2,10000)
```

```python
classification = np.where(data_samples < 0, -1,1)
```

```python
prediction = np.ones(len(data_samples))
```

```python
loss = classification != prediction
```

```python
#plt.ylim(0,4)
#plt.xlim(-0.00005,0.00005)
plt.scatter(data_samples,loss, label = "0-1 loss")
plt.show()
```
